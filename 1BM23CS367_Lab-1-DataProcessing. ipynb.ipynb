```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
```

```python
diabetes = pd.read_csv("diabetes.csv")
```

```python
diabetes.isnull().sum()
```

```python
diabetes.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
diabetes['Gender'] = le.fit_transform(diabetes['Gender'])
diabetes['CLASS'] = le.fit_transform(diabetes['CLASS'])
```

```python
num_cols = [
    'AGE','Urea','Cr','HbA1c','Chol',
    'TG','HDL','LDL','VLDL','BMI'
]

Q1 = diabetes[num_cols].quantile(0.25)
Q3 = diabetes[num_cols].quantile(0.75)
IQR = Q3 - Q1

diabetes = diabetes[~(
    (diabetes[num_cols] < (Q1 - 1.5 * IQR)) |
    (diabetes[num_cols] > (Q3 + 1.5 * IQR))
).any(axis=1)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
diabetes[num_cols] = scaler.fit_transform(diabetes[num_cols])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
diabetes[num_cols] = scaler.fit_transform(diabetes[num_cols])
```

Adult dataset

```python
adult = pd.read_csv("adult.csv")
```

```python
adult.describe()
```

```python
adult.replace('?', np.nan, inplace=True)
adult.isnull().sum()
```

```python
adult.dropna(inplace=True)
```

```python
cat_cols = adult.select_dtypes(include='object').columns
cat_cols
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in cat_cols:
    adult[col] = le.fit_transform(adult[col])
```

```python
num_cols = [
    'age','fnlwgt','educational-num',
    'capital-gain','capital-loss','hours-per-week'
]

Q1 = adult[num_cols].quantile(0.25)
Q3 = adult[num_cols].quantile(0.75)
IQR = Q3 - Q1

adult = adult[~(
    (adult[num_cols] < (Q1 - 1.5 * IQR)) |
    (adult[num_cols] > (Q3 + 1.5 * IQR))
).any(axis=1)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
adult[num_cols] = scaler.fit_transform(adult[num_cols])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
adult[num_cols] = scaler.fit_transform(adult[num_cols])
```
